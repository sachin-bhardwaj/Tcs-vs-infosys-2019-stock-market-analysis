```python
daily_closing_tcs = tcs['Adj Close']
# daily returns
daily_pct_chnge_tcs= daily_closing_tcs.pct_change()
daily_pct_chnge_tcs.fillna(0,inplace=True)
daily_pct_chnge_tcs.head()
```

```python
daily_pct_chnge_tcs.hist(bins=50)
plt.show()
```

```python
# neither gives us profit nor give us loss so useless stock
```

```python
daily_closing_infosys = infosys['Adj Close']
# daily returns
daily_pct_chnge_infosys= daily_closing_infosys.pct_change()
daily_pct_chnge_infosys.fillna(0,inplace=True)
daily_pct_chnge_infosys.head()
```

```python
daily_pct_chnge_infosys.hist(bins=50)
plt.show()
```

```python
# subplot using seaborn
```

```python
import seaborn as sns
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(14,6))
#simple histogram
sns.distplot(daily_pct_chnge_tcs,color="b",ax=axes[0],axlabel='Tcs')
sns.distplot(daily_pct_chnge_infosys,color="r",ax=axes[1],axlabel='Infosys')
```

```python
# Finally we chrck the volatility of stock 
#### Higher the volatility riskier the investment in that stock
```

```python
import numpy as np
```

```python
min_periods = 75
# calculating the volatility
vol = daily_pct_chnge_infosys.rolling(min_periods).std()*np.sqrt(min_periods)
vol.fillna(0,inplace=True)
vol.head()
```

```python
vol.plot(figsize=(12,6))
plt.show()
```

```python
tcs_adj_close_px = tcs['Adj Close']
# Short moving window rolling mean
tcs['42'] = tcs_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
tcs['252'] = tcs_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
tcs[['Adj Close', '42', '252']].plot(title="TCS")

# Show plot
plt.show()

infosys_adj_close_px = infosys['Adj Close']
# Short moving window rolling mean
infosys['42'] = infosys_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
infosys['252'] = infosys_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
infosys[['Adj Close', '42', '252']].plot(title="Infosys")

# Show plot
plt.show()
```
